```python
def same_tree_check(p, q):
  if not p and not q:
    return True
  if (not p and q) or (p and not q):
    return False
  if p.val == q.val:
    return same_tree_check(p.left, q.left) and same_tree_check(p.right, q.right)
  return False
```
