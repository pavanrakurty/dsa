```python
def next_greater_element(nums1, nums2):

  n = len(nums1)
  res = [-1]*n
  comp = tuple(nums2)

  for i in range(n):
    if nums2:
      x = nums2.pop() ## x= 2
    res_num = res[i]  ## res_num = -1
    while nums2 and nums1[i] != x: ##nums1[i] = 1
      res_num = x if x > nums1[i] else res_num #res_num = 3
      x = nums2.pop() #x = 1,  nums2 = [0]
    res[i] = res_num
    nums2 = list(comp)

  return res

nums1 = [2,4]
nums2 = [1,2,3,4]
next_greater_element(nums1, nums2)
```

```python
def next_greater_element_sol2(nums1=None, nums2=None):
  st = []
  next_greater = {}

  for x in nums2:
    while st and st[-1] < x:
      next_greater[st.pop()] = x
    st.append(x)

  for x in st:
    next_greater[x] = -1

  return [next_greater[x] for x in nums1]

next_greater_element_sol2(nums1 = [2,4], nums2 = [1,2,3,4])
```
