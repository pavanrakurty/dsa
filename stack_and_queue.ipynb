```python
#stack -> Last in First out(LIFO)
stk = []

stk.append(1)
stk.append(2)
stk.append(3)
stk.append(4)

stk
```

```python
x = stk.pop()
print(x)
print(stk)
```

```python
##queue (First In First Out)

from collections import deque

q = deque()

q.append(1) ## -> O(1)
q.append(2)
q.append(3)
q.append(4)

q
```

```python
##remove element from the first --> O(1)

x = q.popleft()
print(x)
print(q)
```
