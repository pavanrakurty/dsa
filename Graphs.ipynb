```python
n = 8 ##number of nodes / vertices
A = [[0,1],
     [1,3],
     [0,2],
     [2,4],
     [4,3],
     [4,5],
     [2,6],
     [2,7]
     ]
```

```python
M= []

for i in range(n):
  M.append([0]*n)
  ##creates empty matrix


for u, v in A:
  M[u][v] = 1
  #M[v][u] = 1 ##use this if the graph is bi-directional/undirected

M
```

```python
##adjacency list

from collections import defaultdict

D = defaultdict(list)
for u, v in A:
  D[u].append(v)
  #D[v].append(u) ##if the graph is undirected / bi-directional

print(D)
```

```python
##recursive DFS to traverse through graph

def recursive_dfs(source):
  print(source)

  for nei_node in D[source]:
    seen.add(nei_node)
    recursive_dfs(nei_node)


source = 0
seen = set()
seen.add(source)
recursive_dfs(source)
```

```python
## Iterative DFS
## Time complexity -> O(V + E)

def iterative_dfs(source):
  st = []
  st.append(source)
  seen = set()
  seen.add(source)
  print(source)
  while st:
    x = st.pop()
    for nei_node in D[x]:
      if nei_node not in seen:
        seen.add(nei_node)
        print(nei_node)
      st.append(nei_node)

iterative_dfs(0)
```

```python
##bfs
## Time complexity -> O(V + E)

from collections import deque

source = 0
q = deque()

seen = set()
seen.add(source)
q.append(source)

while q:
  node = q.popleft()
  print(node)

  for nei_node in D[node]:
    if nei_node not in seen:
      seen.add(nei_node)
      q.append(nei_node)
```

```python
class Node:

  def __init__(self, val=None):
    self.val = val
    self.neighbours = []

  def __str__(self):
    return f'Node({self.val})'

  def display(self):
    connections = [node.val for node in self.neighbours]
    return f'{self.val} is connected to {connections}'

  def add_neighbour(self, node):
    if node not in self.neighbours:
      self.neighbours.append(node)


A = Node('A')
B = Node('B')
C = Node('C')
D = Node('D')
E = Node('E')

A.add_neighbour(B)
A.add_neighbour(C)
B.add_neighbour(D)
C.add_neighbour(D)
D.add_neighbour(E)

print(A)
print(A.display())

print(B)
print(B.display())

print(C)
print(C.display())

print(D)
print(D.display())

print(E)
print(E.display())
```
