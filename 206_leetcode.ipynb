```python
class Node:
  def __init__(self, val = None, next =None):
    self.val = val
    self.next = next

  def __str__(self):
    return f'Node({self.val})'
```

```python
A = Node(1)
B = Node(2)
C = Node(3)
D = Node(4)
E = Node(5)

A.next = B
B.next = C
C.next = D
D.next = E
```

```python
def display_list(head):
  curr = head
  while curr:
    print(curr , '->')
    curr = curr.next

display_list(A)
```

```python
def reverse_list(head):

  temp = None

  curr = head

  while curr:
    next_node = curr.next
    curr.next = temp
    temp = curr
    curr = next_node
  display_list(temp)

reverse_list(A)
```
