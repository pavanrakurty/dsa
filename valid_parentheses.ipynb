```python
# Input: s = "()"
# Output: true

# Input: s = "()[]{}"
# Output: true

# Input: s = "(]"
# Output: false

def valid_parenthsis(s):

  ref = {'}':'{',
         ')':'(',
         ']' : '['
         }
  st = []

  for x in s:
    if x not in ref:
      st.append(x)
    else:
      if st:
        left = st.pop()
      else:
        return False
      if not left == ref[x]:
        return False
  if not st:
    return True
  else:
    return False
```

```python
valid_parenthsis('(([{]))')
```
