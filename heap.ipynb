```python
A = [10, 40, 25, 12, 20, 15, 18]

import heapq ##used to create min heap

heapq.heapify(A) ## O(n)

A ##creates min heap
```

```python
heapq.heappush(A, -4)
A

## A is not sorted yet
```

```python
##heap pop (extracts min element)

minn = heapq.heappop(A)

A, minn
```

```python
##heap sort
## time complexity -> n*log(n)
## space complexity -> O(n)

def heap_sort(arr):
  n = len(arr)
  res = [0]*n

  heapq.heapify(arr)

  for i in range(n):
    minn = heapq.heappop(arr)
    res[i] = minn

  return res
```

```python
heap_sort([10, 12, 15, 40, 20, 25, 18 ,-4])
```

```python
## building heap from scratch

## this is O(n log (n)) operation, as it takes log(n) time to push and for n elements it is n log(n)

C = [40, 5, 7,90,10,-1]

n = len(C)
heap = []
for x in C:
  heapq.heappush(heap, x)
  print(heap)
```

```python
##important for leetcode problems
## creating heap using tuples

## let us create min heap with least occured element in list first

N = [5,4,3,2,1,4,3,2,5,6,4,3,5,6,6,7,1,2,2,5,3,3,2,2,2,2,6,6,6,6,6,6,6,6,6]

from collections import Counter

c = Counter(N)
c
```

```python
for k,v in c.items():
  print(k,v)
```

```python
heap = []

for k,v in c.items():
  heapq.heappush(heap, (v, k)) ##as we are trying to sort element with element frequency, we give the frequency as the first object

heap
```
