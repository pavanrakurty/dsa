```python
##binary search iterative method

def binary_search_iterative(A, target):
  n = len(A)
  l = 0
  r = n - 1

  while l<=r:
    mid = l + ((r-l)//2)

    if A[mid] == target:
      return True

    if A[mid] < target:
      l = mid +1
    else:
      r = mid -1

  return False
```

```python
A = [-1,1,2,3,4,5,10]

binary_search_iterative(A, -2)
```
