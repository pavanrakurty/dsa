```python
##fibonacci
## Time complexity -> O(2**n), Space complexity -> O(n)
def F(n):
  if n==0: return 0
  elif n==1: return 1

  return F(n-1)+F(n-2)

F(12)
```

```python
##reverse a linkedlist

class Node:

  def __init__(self, val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)


head = Node(1)
A = Node(2)
B = Node(3)
C = Node(4)

head.next = A
A.next = B
B.next = C
```

```python
def display(head):
  curr = head

  while curr:
    print(curr)
    curr = curr.next

display(head)
```

```python
##print reversed linkedlist using recursion


def reverse(head):
  ##time complexity O(n)
  ##space complexity O(n)

  if not head: ## -> base case
    return

  reverse(head.next)
  print(head)


reverse(head)
```
