```python
class TreeNode:

  def __init__(self, val=None, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(2)
D = TreeNode(3)
E = TreeNode(4)
F = TreeNode(4)
G = TreeNode(3)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F
C.right = G
```

```python
def isSymmetricTree(p):

  def recursive_symmetric_check(p, q):

    if not p and not q:
      return True

    if (not p and q) or (not q and p):
      return False

    if p.val == q.val:
      return recursive_symmetric_check(p.left,q.right) and recursive_symmetric_check(p.right, q.left)

    return False


  return recursive_symmetric_check(p.left, p.right)



isSymmetricTree(A)
```
