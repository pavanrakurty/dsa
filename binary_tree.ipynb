```python
##binary tree

class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
root = TreeNode(1)
A = TreeNode(2)
B = TreeNode(3)
C = TreeNode(4)
D = TreeNode(5)
E = TreeNode(10)

root.left = A
root.right = B
A.left = C
A.right = D
B.left = E

##      1
##   2     3
## 4  5  10
```

```python
##recursvie pre-order Traversal(DFS)
##node -> left -> right
## time complexity -> O(n)
## space complexity -> O(n)

def preorder(root):

  if not root:
    return

  print(root)
  preorder(root.left)
  preorder(root.right)

preorder(root)
```

```python
##recursive inorder traversal
## left -> node -> right

def inorder(root):
  if not root:
    return

  inorder(root.left)
  print(root)
  inorder(root.right)

inorder(root)
```

```python
##recursive post order traversal
## left -> right -> node

def postorder(root):
  if not root:
    return

  postorder(root.left)
  postorder(root.right)
  print(root)

postorder(root)
```

```python
##preorder traversal using loop (not recursion)
##uses stack

def preorder_iterative(root):
  st = [root]

  while st:
    x = st.pop()
    print(x)
    if x.right: st.append(x.right)
    if x.left: st.append(x.left)

preorder_iterative(root)
```

```python
##BFS or Level order Traversal
##uses queue
from collections import deque

def bfs(root):
  q = deque()
  q.append(root)
  while q:
    root = q.popleft()
    print(root)
    if root.left: q.append(root.left)
    if root.right: q.append(root.right)

bfs(root)
```

```python
##check if a value exists
##let us use DFS

def search_val(root, target):
  st = [root]

  while st:
    x = st.pop()
    if target == x.val: return True
    if x.right: st.append(x.right)
    if x.left: st.append(x.left)

  return False

search_val(root, 10)
```

```python
##method 2 -> using recursion

def search_val_recursion(root, target):

  if not root:
    return False

  if root.val == target:
    return True

  return search_val_recursion(root.left, target) or search_val_recursion(root.right, target)

search_val_recursion(root, 10)
```

```python
##binary search tree

A2 = TreeNode(5)
B2 = TreeNode(1)
C2 = TreeNode(8)
D2 = TreeNode(-1)
E2 = TreeNode(3)
F2 = TreeNode(7)
G2 = TreeNode(9)

A2.left = B2
B2.left = D2
B2.right = E2
A2.right = C2
C2.left = F2
C2.right = G2
```

```python
def inorder_pr(root):  ##same inorder function as above

  if not root:
    return

  inorder_pr(root.left)
  print(root)
  inorder_pr(root.right)

inorder_pr(A2)
```

```python
def search_bst(root, target):
  if not root:
    return False

  if root.val == target:
    return True

  if target < root.val:
    return search_bst(root.left, target)
  else:
    return search_bst(root.right, target)

search_bst(A2, 8)
```
