```python
## selection sort
## time complexity ->O(n * n)
## space complexity -> O(1) (in place sort)

def selection_sort(arr):

  n = len(arr)
  for i in range(n-1):
    for j in range(i+1, n):
      if arr[i] > arr[j]:
        arr[i], arr[j] = arr[j], arr[i]
  return arr

A = [-5, 5, 6, -100, 100, -4, 1, 3, 2]

selection_sort(A)
```

```python
## bubble sort
## time complexity ->O(n * n)
## space complexity -> O(1) (in place sort)

def bubble_sort(arr):
  n = len(arr)

  for i in range(n):
    for j in range(n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]

  return arr

A = [-5, 5, 6, -100, 100, -4, 1, 3, 2]

bubble_sort(A)
```

```python
## insertion sort
## time complexity ->O(n * n)
## space complexity -> O(1) (in place sort)

def insertion_sort(arr):
  n = len(arr)

  for i in range(1,n):
    j = i
    while j>=1 and arr[j] < arr[j-1]:
      arr[j] , arr[j-1] = arr[j-1], arr[j]
      j -= 1
  return arr

A = [-5, 5, 6, -100, 100, -4, 1, 3, 2]

insertion_sort(A)
```

```python
##merge sort
##time complexity -> O(n* log n)
## space complexity -> O(n)

def merge_sort(arr):

  n = len(arr)
  if n == 1:
    return arr

  m = n // 2

  L = arr[:m]
  R = arr[m:]

  L = merge_sort(L)
  R = merge_sort(R)

  l_len = len(L)
  r_len = len(R)

  i = l = r = 0
  sorted_arr = [0] * n

  while l < l_len and r < r_len:
    if L[l] < R[r]:
      sorted_arr[i] = L[l]
      l += 1
    else:
      sorted_arr[i] = R[r]
      r += 1
    i += 1

  while l < l_len:
    sorted_arr[i] = L[l]
    l += 1
    i += 1

  while r < r_len:
    sorted_arr[i] = R[r]
    r += 1
    i += 1

  return sorted_arr

A = [-5, 5, 6, -100, 100, -4, 1, 3, 2]

merge_sort(A)
```

```python
##quick sort
## time complexity -> O(n log n) on average, O(n^2) worst case
## space complexity -> O(log n) average, O(n) worst case (for recursion stack)

def partition(A, l, h):
  # This function returns the pivot's final position.
  # The pivot is chosen as the first element (A[l]).
  # Elements smaller than the pivot are moved to its left, and larger to its right.

  i, j = l, h
  pivot = A[l] # Use A[l] as pivot, not A[i] directly inside loop

  while (i < j):
    # Increment i until an element greater than the pivot is found
    while i <= h and A[i] <= pivot:
      i += 1
    # Decrement j until an element smaller than or equal to the pivot is found
    while A[j] > pivot:
      j -= 1

    if (i < j):
      A[i], A[j] = A[j], A[i] # Swap elements if i is still less than j

  # Place the pivot in its correct sorted position
  A[l], A[j] = A[j], A[l]
  return j # Return the final position of the pivot


def quick_sort(A, l, h):
  # Base case: if the sub-array has 0 or 1 elements, it's already sorted
  if l < h:
    # Partition the array and get the pivot's final position
    partition_idx = partition(A, l, h)

    # Recursively sort the sub-array to the left of the pivot
    quick_sort(A, l, partition_idx - 1)

    # Recursively sort the sub-array to the right of the pivot
    quick_sort(A, partition_idx + 1, h)
  return A


A = [-5, 5, 6, -100, 100, -4, 1, 3, 2]
# A = [40, 30, 50, 60, 70, 20]

# Call quick_sort with the array and its full range
print(quick_sort(A, 0, len(A)-1))
```
