```python
def backspaceCompare(s, t):
  st1, st2 = [], []

  for x in s:
    if st1 and x == '#':
      st1.pop()
    if x != '#':
      st1.append(x)

  for x in t:
    if st2 and x == '#':
      st2.pop()
    if x != '#':
      st2.append(x)
  print(st1, st2)

  return st1 == st2

backspaceCompare("c#b", t = "b")
```
