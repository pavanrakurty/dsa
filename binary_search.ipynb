```python
##naive binary search
##time complexity -> O(n)

A = [-3,-2,-1,0,1,2,3]
target = 3

if target in A:
  print(True)
```

```python
##traditional binary search
## time complexity -> O(log n)
## space complexity -> O(1)

def traidional_bs(arr, target):

  N = len(arr)
  L = 0
  R = N-1

  while L <= R:
    M = L + ((R-L)//2)

    if arr[M] == target:
      return True
    elif arr[M] > target:
      R = M-1
    else:
      L = M+1

  return False


print(traidional_bs(A, -3))
```

```python
##condition based search

B = [False, False, False, True, True, True, True]

def condition_bs(arr):

  N = len(arr)
  L = 0
  R = N

  while L < R:
    M = L + ((R-L)//2)

    if B[M]:
      R = M
    else:
      L = M+1

  return L

condition_bs(B)


```
